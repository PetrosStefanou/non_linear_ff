```python
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.patches import Circle
```

```python
#output settings
np.set_printoptions(precision=2)
%matplotlib qt
```

```python
#the grid

Nt = 100
Nx = 100

T = 5.
L = 6.3

t = np.linspace(0, T, Nt)
x = np.linspace(0, L, Nx)

dt, dx = T/Nt, L/Nx
```

```python
def F(u):

    f = u**2/2

    return f
```

```python
u = np.zeros((Nt, Nx))

u[0,:] = np.sin(x)

# print(u)

for n in range(0, len(t)-1):
    
    # print(np.min(u[n])*dt/dx <= 1) #and u[n,j]*dt/dx >= 0)
    
    for j in range (0, len(x)-1):

        if u[n,j] < 0:

            u[n+1,j] =u[n,j] - dt/dx*(F(u[n,j+1]) - F(u[n,j]))

            print('-')

        else: 
            
            u[n+1,j] =u[n,j] - dt/dx*(F(u[n,j]) - F(u[n,j-1]))
        
            u[n+1,len(x)-1] =u[n,len(x)-1] - dt/dx*(F(u[n,len(x)-1]) - F(u[n,len(x)-1-1]))    

        print('+')    
        
```

```python
fig,ax = plt.subplots()

for n in range(0, Nt, int(Nt/10)):
    
    ax.plot(x, u[n, :], label='t = {:1.2}'.format(t[n]))

ax.set_ylim(-20,20)
ax.legend()
```
